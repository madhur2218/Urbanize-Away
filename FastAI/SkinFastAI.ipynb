```python

from PIL.Image import ROTATE_90,ROTATE_180
SIZE=128
for fn in allfiles:
  img = PILImage.create(fn)
  methods = [ROTATE_180]
  for method in methods:
    tfm = Resize(SIZE, method=method)
    tfm(img).save(f'/content/drive/MyDrive/skin_burn/Train/degree1/-{str(method)}-{fn.name}')
```
